```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from simpson_gss import *
```

```python
gss = pd.read_hdf('gss_simpson', 'gss')
```

```python
def make_plot(title, ylabel):
    visualize(series_all, table)
    plt.title('')

    
    plt.title(title, loc='left', fontdict=dict(fontsize=14))
    plt.ylabel(ylabel)
    plt.xlabel(xvarname.capitalize())

    x = y = 1.02
    plt.legend(title='Birth decade', bbox_to_anchor=(x, y), loc='upper left', ncol=1)   
    plt.tight_layout()

    filename = '%s_vs_%s_by_%s.jpg' % (yvarname, xvarname, gvarname)
    plt.savefig(filename)
```

## racmar vs age

```python
xvarname = 'age'
yvarname = 'racmar'
gvarname = 'cohort10'
yvalue = 1

run_subgroups(gss, xvarname, yvarname, gvarname, yvalue)
```

```python
series_all, table = summarize(gss, xvarname, yvarname, gvarname, yvalue)
table.drop([1890, 1980], axis=1, inplace=True)
#table.sort_index(axis=1, ascending=False, inplace=True)
table
```

```python
title = """Do you think there should be laws against marriages between 
(Negroes/Blacks/African-Americans) and whites?
"""

ylabel = 'Fraction saying yes'
make_plot(title, ylabel)
```
